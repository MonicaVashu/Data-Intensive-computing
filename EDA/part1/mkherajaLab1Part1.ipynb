# Problem 1

The second set of data is a random sample from poisson's distribution. Hence, everytime we execute the code, a different set of values are assigned to sales2 and hence a different graph is obtained.

```r
sales1<-c(12,14,16,29,30,45,19,20,16, 19, 34, 20)
sales2<-rpois(12,34)  # random numbers, Poisson distribution, mean at 34, 12 numbers

par(bg="cornsilk")
plot(sales1, col="blue", type="o", ylim=c(10,50), xlab="Month", ylab="Sales" )
title(main="Sales by Month")

lines(sales2, type="o", pch=22, lty=2, col="red")
grid(nx=NA, ny=NULL)
legend("topright", inset=.05, c("Sales1","Sales2"), fill=c("blue","red"))
```

# Problem 4

Comparison of prices each day of the two giants- Apple and facebook. Here also, the data files are choosen while running the code. Different types of points can be used here, with variety of combination of point and lines specified by "type", different colours, etc. 
The graph shows that facebook beats apple most of the time except for one time when apple's price exceeds that of facebook's. 

```r
fb1<-read.csv(file.choose())
aapl1<-read.csv(file.choose())

par(bg="cornsilk")

plot(aapl1$Adj.Close, col="blue", type="o", ylim=c(150,200), xlab="Days", ylab="Price" )
lines(fb1$Adj.Close, type="o", pch=22, lty=2, col="red")
legend("topright", inset=.05, c("Apple","Facebook"), fill=c("blue","red"), horiz=TRUE)

# Just study the distribution of the adjusted close of the stock price of Apple.
hist(aapl1$Adj.Close, col=rainbow(8))
```

# Problem 6

Library ggmap is used to plot the desired location on the map with various styles.

```r
library("ggmap")
library("maptools")
library(maps)

register_google(key = "") 

visited <- c("Buffalo", "Chennai", "London", "Melbourne", "Mumbai", "Tokyo", "Moscow", "Rome", "Cape Town", "Brazil", "Sitka")
ll.visited <- geocode(visited)
visit.x <- ll.visited$lon
visit.y <- ll.visited$lat
map("world", fill=TRUE, col="white", bg="lightblue", ylim=c(-60, 90), mar=c(0,0,0,0))
points(visit.x,visit.y, col="red", pch=36)
```

```r
library("ggmap")
library("maptools")
library(maps)

visited <- c("SFO", "New York", "Buffalo", "Dallas, TX", "Austin")
ll.visited <- geocode(visited)
visit.x <- ll.visited$lon
visit.y <- ll.visited$lat
map("state", fill=TRUE, col=rainbow(50), bg="white", mar=c(0,0,0,0))
points(visit.x,visit.y, col="black", pch=36)
```
